```python
from sentence_transformers import SentenceTransformer, util
model = SentenceTransformer('multi-qa-MiniLM-L6-cos-v1')
```

```python
import pandas as pd
df_postive = pd.read_csv("./postive.csv")
df_postive=df_postive.drop_duplicates(subset='q_text')
```

```python
q_id_list=df_postive['q_id'].values.tolist()
q_text_list=df_postive['q_text'].values.tolist()
#生成字典
dict={}
for i in range(0,len(q_id_list)):
    dict.update({q_text_list[i]:q_id_list[i]})
#获取所有语料
all_list=[]
for i in q_text_list:
    temp_list=[i]
    all_list.append(temp_list)
```

```python
passage_embedding = model.encode(q_text_list,device='cuda',show_progress_bar=True)
```

```python
from tqdm import tqdm
id_matix=[]
query_embedding = model.encode(q_text_list, device='cuda',show_progress_bar=True)
cos_scores = util.semantic_search(query_embedding, passage_embedding,top_k=2,query_chunk_size=100000,corpus_chunk_size=100000000)
```

```python
for index,i in enumerate(cos_scores):
    id_matix.append(cos_scores[index][1])
id_matix
```
