```python
import pandas as pd
df_negtive=pd.read_csv('./neutral.csv')
df_negtive
```

```python
import random
#设定一个随机数生成器
#生成一个列表答案列表
negitive_list=[]
for i in df_negtive.index:
    temp=random.randint(0,32651)
    negitive_list.append(df_negtive.iloc[temp].a_text)
```

```python
df_negtive['a_text']=negitive_list
df_negtive['flag']=-2
```

```python
df_negtive = df_negtive.drop(df_negtive.columns[0], axis=1)
df_negtive=df_negtive.drop(['q_id','u_id','a_id','r','cos'], axis=1)    
df_negtive.to_csv('negtive.csv',index=True)
```

```python
df_neutrl=pd.read_csv('./neutral.csv')
df_neutrl = df_neutrl.drop(df_neutrl.columns[0], axis=1)
df_neutrl=df_neutrl.drop(['q_id','u_id','a_id','r','cos'], axis=1)
df_neutrl.flag=-1
df_neutrl
```

```python
df_postive=pd.read_csv('./postive.csv')
df_postive = df_postive.drop(df_postive.columns[0], axis=1)
df_postive=df_postive.drop(['q_id','u_id','a_id','r'], axis=1) 
df_postive
```

```python
df=pd.concat([df_postive,df_neutrl])
df=pd.concat([df,df_negtive])
```

```python
df.to_csv('stackall.csv',index=True)
df
```
